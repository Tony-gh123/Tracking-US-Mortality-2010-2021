```python
%%sql
SELECT * FROM table3
WHERE Location = 'USA'
AND (
    [Risk Name] LIKE '%High body-mass index%'
    OR [Risk Name] LIKE '%Particulate matter pollution%'
    OR [Risk Name] LIKE '%Smoking%'
    OR [Risk Name] LIKE '%Low physical activity%'
    OR [Risk Name] LIKE '%Diet low in fibre%'
    OR [Risk Name] LIKE '%Diet low in whole grains%'
    OR [Risk Name] LIKE '%Diet high in sugar-sweetened beverages%'
    OR [Risk Name] LIKE '%Diet high in processed meat%'
    OR [Risk Name] LIKE '%High alcohol use%'
    OR [Risk Name] LIKE '%Diet low in fruits%'
    OR [Risk Name] LIKE '%Diet high in red meat%'
    OR [Risk Name] LIKE '%Diet low in calcium%'
    OR [Risk Name] LIKE '%High fasting plasma glucose%'
    OR [Risk Name] LIKE '%Diet low in milk%'
    OR [Risk Name] LIKE '%Diet low in vegetables%'
    OR [Risk Name] LIKE '%Non-optimal temperature%'
    OR [Risk Name] LIKE '%High LDL cholesterol%'
    OR [Risk Name] LIKE '%High systolic blood pressure%'
    OR [Risk Name] LIKE '%Kidney dysfunction%'
) limit 2;
```

```python
%%sql risk_name_changes <<
select [Risk Name], [Percentage change 2010-2021 Male], [Percentage change 2010-2021 Female], [Percentage change 2010-2021 Both], [Location]
from table3
where Location = 'USA'
AND (
    [Risk Name] LIKE '%High body-mass index%'
    OR [Risk Name] LIKE '%Particulate matter pollution%'
    OR [Risk Name] LIKE '%Smoking%'
    OR [Risk Name] LIKE '%Low physical activity%'
    OR [Risk Name] LIKE '%Diet low in fibre%'
    OR [Risk Name] LIKE '%Diet low in whole grains%'
    OR [Risk Name] LIKE '%Diet high in sugar-sweetened beverages%'
    OR [Risk Name] LIKE '%Diet high in processed meat%'
    OR [Risk Name] LIKE '%High alcohol use%'
    OR [Risk Name] LIKE '%Diet low in fruits%'
    OR [Risk Name] LIKE '%Diet high in red meat%'
    OR [Risk Name] LIKE '%Diet low in calcium%'
    OR [Risk Name] LIKE '%High fasting plasma glucose%'
    OR [Risk Name] LIKE '%Diet low in milk%'
    OR [Risk Name] LIKE '%Diet low in vegetables%'
    OR [Risk Name] LIKE '%Non-optimal temperature%'
    OR [Risk Name] LIKE '%High LDL cholesterol%'
    OR [Risk Name] LIKE '%High systolic blood pressure%'
    OR [Risk Name] LIKE '%Kidney dysfunction%'
)
```

```python
df = risk_name_changes.DataFrame()

for col in df.columns:
    df[col] = df[col].map(lambda x: str(x).replace('\n', ' ') if pd.notnull(x) else x)

df_clean = df[df["Risk Name"].astype(str).str.strip().ne("")]

for col in [
    "Percentage change 2010-2021 Male",
    "Percentage change 2010-2021 Female",
    "Percentage change 2010-2021 Both"
]:
    df_clean[col] = df_clean[col].str.extract(r'([-+]?\d+·\d+)')

print(df_clean.to_markdown(index=False))
```

```python
# Rename columns so they're consistent for melting
df_clean = df_clean.rename(columns={
    "Percentage change 2010-2021 Male": "Male",
    "Percentage change 2010-2021 Female": "Female",
    "Percentage change 2010-2021 Both": "Both"
})

# Only melt the sex columns
df_melted = df_clean.melt(
    id_vars="Risk Name",
    value_vars=["Male", "Female", "Both"],
    var_name="Sex",
    value_name="Percentage Change"
)

# Convert 'Percentage Change' to float
df_melted["Percentage Change"] = (
    df_melted["Percentage Change"]
    .str.replace("·", ".", regex=False)
    .astype(float)
)

# Order risks by "Both" percentage change for readability
order = df_clean.sort_values("Both")["Risk Name"]

# Plot
plt.figure(figsize=(10, 8))
sns.barplot(
    data=df_melted,
    y="Risk Name",
    x="Percentage Change",
    hue="Sex",
    order=order,
    palette="Set2"
)

plt.axvline(0, color="black", linewidth=0.8)
plt.title("U.S. Risk Factor Exposure Trends, 2010–2021 (% Change in SEV)", fontsize=14, weight="bold", pad=20)
plt.xlabel("Percentage Change (%)")
plt.ylabel("Risk Factor")
plt.legend(title="Sex", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```
